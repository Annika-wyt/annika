```python
import numpy as np
import cv2
import rosbag
import rospy
```

```python
################
### ALL BAGS ###
################

BagPaths = {
    1: '/home/annika/ITRL/kth_thesis/rosbag/2024-01-22/working_epnp_with_EKF.bag',
    2: ' '
    }

Bag = BagPaths[BAGNUMBER]
```

```python
#########################
### GET DATA FROM BAG ###
#########################

Svea5 = []
# MapAruco =
# Open the bag file
map_aruco_last_timestamp = None
aruco_camera_last_timestamp = None
duration_threshold = rospy.Duration(0.01) 
with rosbag.Bag(Bag, 'r') as bag:
    # Iterate through messages in the bag
    for topic, msg, t in bag.read_messages():
        if topic == '/qualisys/svea5/pose':
            Svea5.append([msg.pose.position.y - 0.06, -msg.pose.position.x + 0.06])
        if topic == '/tf' and "MapAruco" in msg.transforms[0].child_frame_id:
            map_aruco_last_timestamp = t
            print("map")
            if aruco_camera_last_timestamp and (map_aruco_last_timestamp - aruco_camera_last_timestamp) < duration_threshold:
                print("Map", msg.transforms[0].transform.translation.x, msg.transforms[0].transform.translation.y)

        elif topic == '/tf' and "arucoCamera" in msg.transforms[0].child_frame_id:
            aruco_camera_last_timestamp = t
            print("camera")
            if map_aruco_last_timestamp and (aruco_camera_last_timestamp - map_aruco_last_timestamp) < duration_threshold:
                print("Camera", msg.transforms[0].transform.translation.x, msg.transforms[0].transform.translation.y)

Svea5 = np.array(Svea5)
```
