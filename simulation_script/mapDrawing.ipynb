```python
%matplotlib widget
fig, ax = plt.subplots()

# Generate some data
width = 4.0
dotted_line_skrink = 0.4
shrink = 0.3
x = np.linspace(0, width, 2)
t = np.ones(2)*width
# Create the plot
# plt.plot(t, x, label='x Wave')
# plt.plot(x, t, label='y Wave')
ax.hlines(width/2 - 0.5, width/2 + 0.5 + shrink, width, colors="blue")
ax.hlines(width/2 + 0.5, width/2 + 0.5 + shrink, width, colors="blue")
ax.hlines(width/2 - 0.5, 0.0, width/2 - 0.5 - shrink, colors="blue")
ax.hlines(width/2 + 0.5, 0.0, width/2 - 0.5 - shrink, colors="blue")

ax.vlines(width/2 - 0.5, width/2 + 0.5 + shrink, width, colors="blue")
ax.vlines(width/2 + 0.5, width/2 + 0.5 + shrink, width, colors="blue")
ax.vlines(width/2 - 0.5, 0.0, width/2 - 0.5 - shrink, colors="blue")
ax.vlines(width/2 + 0.5, 0.0, width/2 - 0.5 - shrink, colors="blue")

ax.plot((2.5, 2.8), (2.8,2.5), linestyle='-', color='blue')
ax.plot((2.5, 2.8), (1.2,1.5), linestyle='-', color='blue')
ax.plot((1.2, 1.5), (1.5,1.2), linestyle='-', color='blue')
ax.plot((1.5, 1.2), (2.8,2.5), linestyle='-', color='blue')

ax.hlines(width/2, width/2 + 0.5 + dotted_line_skrink, width, linestyles="dashed")
ax.vlines(width/2, width/2 + 0.5 + dotted_line_skrink, width, linestyles="dashed")
ax.hlines(width/2, 0.0, width/2 - 0.5 - dotted_line_skrink, linestyles="dashed")
ax.vlines(width/2, 0.0, width/2 - 0.5 - dotted_line_skrink, linestyles="dashed")

ax.set_xlim(0.0, width)
ax.set_ylim(0.0, width)
# ax.set_xlabel("meter")
# ax.set_ylabel("meter")
ax.fill_between(x, t, color='skyblue', alpha=1)  # Adjust alpha for transparency
centers = [(3.2 + 0.4 , 3.0 - 0.35), 
           (3.2 + 0.4 , 1.0 + 0.35), 
           (0.8 - 0.4, 3.0 - 0.35), 
           (0.8 - 0.4, 1.0 + 0.35), 
           (3.0 - 0.35, 3.2 + 0.4), 
           (1.0 + 0.35, 3.2 + 0.4), 
           (3.0 - 0.35, 0.8 - 0.4), 
           (1.0 + 0.35, 0.8 - 0.4), 

           (width/2, width/2 - 1.0),
           (width/2, width/2 + 1.0),
           (width/2 - 1.0, width/2),
           (width/2 + 1.0, width/2)]
dimension = 0.14
small_dimension = 0.05
for idx, center in enumerate(centers):
    rect = patches.Rectangle((center[0] - dimension/2, center[1] - dimension/2), dimension, dimension, rotation_point="xy",linewidth=1, edgecolor='grey', facecolor=None, alpha=0.5)
    ax.add_patch(rect)
    rect = patches.Rectangle((center[0] - small_dimension/2, center[1] - small_dimension/2), small_dimension, small_dimension, rotation_point="xy",linewidth=1, edgecolor='grey', facecolor="grey", alpha=0.5)
    ax.add_patch(rect)
    ax.text(center[0]-0.1, center[1]-0.2, idx+1, color='black', ha='center', va='bottom')
    print("[", center[0]-2.0, ",", center[1]-2.0, ",", 0, "],")
svea_dimension = (0.55, 0.3)
rect = patches.Rectangle((3.25 - svea_dimension[0]/2, 2.25 - svea_dimension[1]/2), svea_dimension[0], svea_dimension[1], rotation_point="xy",linewidth=1, edgecolor='green', facecolor="green", alpha=0.2)
ax.add_patch(rect)
rect = patches.Rectangle((2.4 - svea_dimension[0]/2, 2.65 - svea_dimension[1]/2), svea_dimension[0], svea_dimension[1], angle=-45, rotation_point="xy",linewidth=1, edgecolor='green', facecolor="green", alpha=0.4)
ax.add_patch(rect)
rect = patches.Rectangle((2.25 - svea_dimension[1]/2, 3.25 - svea_dimension[0]/2), svea_dimension[1], svea_dimension[0], rotation_point="xy",linewidth=1, edgecolor='green', facecolor="green", alpha=0.6)
ax.add_patch(rect)
# Add labels and show the plot
ax.set_aspect('equal', adjustable='box')

plt.show()
```
