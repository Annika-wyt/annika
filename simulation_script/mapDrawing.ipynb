```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
%matplotlib widget
fig, ax = plt.subplots()

# Generate some data
width = 2.0
dotted_line_skrink = 0.4
shrink = 0.3
x = np.linspace(-width, width, 2)
# Create the plot
# plt.plot(t, x, label='x Wave')
# plt.plot(x, t, label='y Wave')
ax.hlines(0.5, -width, -0.5 - shrink, colors="blue")
ax.hlines(-0.5, -width, -0.5- shrink, colors="blue")
ax.hlines(-0.5, 0.5 + shrink, width, colors="blue")
ax.hlines(0.5, 0.5 + shrink, width, colors="blue")

ax.vlines(-0.5, 0.5 + shrink, width, colors="blue")
ax.vlines(0.5,  0.5 + shrink, width, colors="blue")
ax.vlines(-0.5, -0.5 - shrink, -width, colors="blue")
ax.vlines(0.5,  -0.5 - shrink, -width, colors="blue")

ax.plot((0.5, 0.8), (0.8,0.5), linestyle='-', color='blue')
ax.plot((0.5, 0.8), (-0.8,-0.5), linestyle='-', color='blue')
ax.plot((-0.8, -0.5), (-0.5,-0.8), linestyle='-', color='blue')
ax.plot((-0.5, -0.8), (0.8,0.5), linestyle='-', color='blue')

ax.hlines(0.0, 0.5 + dotted_line_skrink, width, linestyles="dashed")
ax.vlines(0.0, 0.5 + dotted_line_skrink, width, linestyles="dashed")
ax.hlines(0.0, -width, -0.5 - dotted_line_skrink, linestyles="dashed")
ax.vlines(0.0, -width, -0.5 - dotted_line_skrink, linestyles="dashed")

ax.set_xlim(-width, width)
ax.set_ylim(-width, width)
ax.set_xlabel("meter")
ax.set_ylabel("meter")
ax.fill_between(x, -2, 2, color='skyblue', alpha=1)  # Adjust alpha for transparency
# # centers = [(3.2 + 0.4 , 3.0 - 0.25), 
# #            (3.2 + 0.4 , 1.0 + 0.25), 
# #            (0.8 - 0.4, 3.0 -  0.25), 
# #            (0.8 - 0.4, 1.0 +  0.25), 
# #            (3.0 - 0.25, 3.2 + 0.4), 
# #            (3.0 - 0.25, 0.8 - 0.4), 
# #            (1.0 + 0.25, 3.2 + 0.4), 
# #            (1.0 + 0.25, 0.8 - 0.4), 

# #            (width/2, width/2 - 1.0),
# #            (width/2, width/2 + 1.0),
# #            (width/2 - 1.0, width/2),
# #            (width/2 + 1.0, width/2)]
centers = [(1.0,  1.4 - 0.7),
           (-1.0, 1.6 - 0.7),
           (0.6,  1.6 - 0.5),
           (-0.6, 1.8 - 0.5)]
landmark_idx = [11,12,14,15]
dimension = 0.14
for idx, center in enumerate(centers):
    rect = patches.Rectangle((center[0] - dimension/2, center[1] - dimension/2), dimension, dimension, rotation_point="xy",linewidth=1, edgecolor='grey', facecolor=None, alpha=0.5)
    ax.add_patch(rect)
    ax.text(center[0]-0.1, center[1]-0.2, landmark_idx[idx], color='black', ha='center', va='bottom')
    print("[", center[0]-2.0, ",", center[1]-2.0, ",", 0, "],")

# # Add labels and show the plot
ax.set_aspect('equal', adjustable='box')

plt.show()
```
